
#### Datos de Participación Electoral:

Posteriormente eliminamos los campos vacíos y normalizamos los datos para tratar las mismas magnitudes. Además, también formamos una nueva estructura de datos que consiste en los votos obtenidos por los tres principales grupos: demócratas, republicanos y "otros" (el resto de partidos minoritarios), por cada candidatura. Finalmente, unificamos los datos del 2000 al 2016 con los datos de 2020.

```r
#Sustituimos los campos vacíos

df_votes_1$party <- as.character(df_votes_1$party)
df_votes_1$party[is.na(df_votes_1$party)] <- "Otros"
df_votes_1$party[df_votes_1$party != 'democrat' & df_votes_1$party != 'republican'] <- "Otros"
df_votes_1$candidatevotes[is.na(df_votes_1$candidatevotes)] <- 0

#Normalizamos para los votos de 2020
df_votes_2$party <- as.character(df_votes_2$party)
df_votes_2$party[is.na(df_votes_2$party)] <- "Otros"
df_votes_2$party[df_votes_2$party != 'DEM' & df_votes_2$party != 'REP'] <- "Otros"
df_votes_2$party[df_votes_2$party == 'DEM'] <- "democrat"
df_votes_2$party[df_votes_2$party == 'REP'] <- "republican"
df_votes_2$total_votes[is.na(df_votes_2$total_votes)] <- 0
```

```r
#Agrupamos los votos por año y partido

new_votes <- df_votes_1 %>% group_by(year, party) %>% summarise(votes = sum(candidatevotes))

#Añadimos columna de año y agrupamos por partido
df_votes_2$year <- as.integer(2020)
new_votes2 <- df_votes_2 %>% group_by(year, party) %>% summarise(votes = sum(total_votes))


#Reordenamos dataset 2 y unimos los valores de ambos dataset
new_votes2 <- new_votes2[, c(1,3,2)]
total_votes <- rbind(new_votes, new_votes2)
```

### Análisis Descriptivo Numérico:

Para el análisis descriptivo numérico se han estudiado las medidas de tendencia central así como las medidas de dispersión para poder entender a grandes rasgos el carácter de los datos.

A continuación, se calcularán las medidas de tendencia central y dispersión. No se aplica summary directamente, se crea una función que devuelve una tabla con toda la información que necesitamos. Además, hemos planteado un estudio separado por partidos y a su vez, otro estudio basado en el total de los votos.

```r
#Función para el cálculo de medidas
mes <- function(data, gdp_values){
    #Tendencia central y dispersión
    mean <- mean(data)
    median <- median(data)
    min <- min(data)
    max <- max(data)
    sd <- sd(data) #Desviación típica
    cv <- sd/mean*100 #Coeficientes de variación
    q1 <- summary(data)['1st Qu.']
    q2 <- summary(data)['Median']
    q3 <- summary(data)['3rd Qu.']
    
    #Tabla dispersión y tendencia
    mes_sol <- cbind(min = min, median = median, mean = mean, max = max, Q1 = q1, Q2 = q2, Q3= q3, std = round(sd,2), cv = round(cv,2))

    return(mes_sol)
}
```

```r
#Total de votos por partido 
dem_vot <- total_votes[total_votes$party == 'democrat',] 
other_vot <- total_votes[total_votes$party == 'Otros',]
rep_vot <- total_votes[total_votes$party == 'republican',]

#Tablas
mes_dem <- mes(dem_vot$votes, gdp_percapita$values)
mes_other <- mes(other_vot$votes, gdp_percapita$values)
mes_rep <- mes(rep_vot$votes, gdp_percapita$values)

#Unión de tablas
mes_party <- cbind( party = c('democrat', 'republican', 'Others'), rbind(mes_dem, mes_rep, mes_other))

mes_party
```

```r
#Suma de votos sin distinción por partido
total_sum <- total_votes %>% group_by(year) %>% summarise(votes = sum(votes))

#Tendencia central y dispersión de la cantidad de votos
mes_vot <- mes(total_sum$votes)
mes_vot
```

```r
#Tendencia central y dispersión del GDP
mes_gdp <- mes(gdp_percapita$values)
mes_gdp
```
